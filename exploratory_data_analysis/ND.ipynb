# ND 
To explore higher dimensional data and relationship betweeen data attributes, load the file `Diabetes.csv`

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
df_original = pd.read_csv("../data/ND/Diabetes.csv")
df_original.head()
```

```python
cols = [c for c in df_original.columns if c not in ["Pregnancies", "Outcome"]]
df = df_original.copy()
df[cols] = df[cols].replace({0:np.NAN})
df.head()
```

```python
df.info()
```

```python
df.describe()
```

# Scatter Matrix
This one is a useful one liner... but note that it only works with numeric data. If you want to include numerical data in there you should convert the categories into numerical variables

```python
pd.plotting.scatter_matrix(df, figsize=(16,7))
```

```python
df2 = df.dropna()
colors = df2["Outcome"].map(lambda x: "#44d9ff" if x else "#f95b4a")
pd.plotting.scatter_matrix(df2, figsize=(16, 6), color=colors)
```

# 2D Histograms
For the rest of this section we are goind to use the data from height_weights.csv (has more data than diabetes dataset)

```python
df2 = pd.read_csv("../data/ND/height_weight.csv")
df2.info()
df2.describe()
```

```python
plt.figure(figsize=(16, 6))
plt.hist2d(df2["height"], df2["weight"], bins=20, cmap="magma")
plt.xlabel("Height")
plt.ylabel("Weight")
```

# Contour Plots 

```python
hist, xedge, yedge = np.histogram2d(df2["height"], df2["weight"], bins=20)
xcenter = 0.5*(xedge[1:] + xedge[:-1])
ycenter = 0.5*(yedge[1:] + yedge[:-1])

plt.figure(figsize=(16, 6))
plt.contour(xcenter, ycenter, hist, levels=4)
plt.xlabel("Height")
plt.ylabel("Weight")
```

# KDE Plots

```python
plt.figure(figsize=(16, 6))
sns.kdeplot(df2["height"], df2["weight"], cmap="viridis", bw=(2, 20))
plt.hist2d(df2["height"], df2["weight"], bins=20, cmap="magma", alpha=0.3)
```

```python
plt.figure(figsize=(16, 6))
sns.kdeplot(df2["height"], df2["weight"], cmap="magma", shade=True)
```

# In Defense of Simplicity 
sometimes we dont need to go fancy. By pausing, taking stock of what we want to see, a scatter plot is normally fairly informative and very fast to plot.

```python
m = df2["sex"] == 1 
plt.figure(figsize=(10, 6))
plt.scatter(df2.loc[m, "height"], df2.loc[m, "weight"], c="#16c6f7", s=1, label="Male")
plt.scatter(df2.loc[~m, "height"], df2.loc[~m, "weight"], c="#ff8b87", s=1, label="Female")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.legend()
```

# Treating points with probability 


```python
params = ["height", "weight"]
male = df2.loc[m, params].values
female = df2.loc[~m, params].values
male.shape
```

```python
from chainconsumer import ChainConsumer
c = ChainConsumer()
c.add_chain(male, parameters=params, name="Male", kde=1.0, color="b")
c.add_chain(female, parameters=params, name="Female", kde=1.0, color="r")
c.configure(contour_labels="confidence", usetex=False)
c.plotter.plot(figsize=2.0)
```

```python
c.plotter.plot_summary(figsize=2.0);
```

Remember, when you're visualising, start simple and add complexity as the data seems to indicate. No point wasting your time getting in real deep when its not needed.
